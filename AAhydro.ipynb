```python
AA=input()
AA=AA.replace(' ', '')
AA=list(AA)
hyd=[]

for acid in AA:
    if acid=='A':
        hyd.append(1.8)
    elif acid=='C':
        hyd.append(2.5)
    elif acid=='D':
        hyd.append(-3.5)
    elif acid=='E':
        hyd.append(-3.5)
    elif acid=='F':
        hyd.append(2.8)
    elif acid=='G':
        hyd.append(-0.4)
    elif acid=='H':
        hyd.append(-3.2)
    elif acid=='I':
        hyd.append(4.5)
    elif acid=='K':
        hyd.append(-3.9)
    elif acid=='L':
        hyd.append(3.8)
    elif acid=='M':
        hyd.append(1.9)
    elif acid=='N':
        hyd.append(-3.5)
    elif acid=='P':
        hyd.append(-1.6)
    elif acid=='Q':
        hyd.append(-3.5)
    elif acid=='R':
        hyd.append(-4.5)
    elif acid=='S':
        hyd.append(-0.8)
    elif acid=='T':
        hyd.append(-0.7)
    elif acid=='V':
        hyd.append(4.2)
    elif acid=='W':
        hyd.append(-0.9)
    elif acid=='Y':
        hyd.append(-1.3)
    else:
        print ('error')
i=0
hydra=[]
while i<len(hyd):
    j=-7
    hydra.append(0)
    while j<8:
        if i+j<0 or i+j>=len(hyd):
            hydra[i]+=0
        else:
            hydra[i]+=hyd[i+j]
        j+=1
    i+=1

import matplotlib.pyplot as plt

plt.plot(hydra)
plt.show()

#i=0
#hydra=[]
#while i<len(hyd):
#    j=-10
#    hydra.append(0)
#    while j<11:
#        if i+j<0 or i+j>=len(hyd):
#            hydra[i]+=0
#        else:
#            hydra[i]+=hyd[i+j]
#        j+=1
#    i+=1
#    
#plt.plot(hydra)
#plt.show()
```
